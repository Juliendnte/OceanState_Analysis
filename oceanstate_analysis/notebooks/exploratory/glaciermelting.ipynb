```python
from analysis.preprocessing import load_and_clean_glaciers_data
```

```python
df = load_and_clean_glaciers_data()
df
```

```python
# Évolution du bilan de masse cumulée des glaciers
import plotly.express as px

fig = px.line(
    df,
    x="Year",
    y="Mean cumulative mass balance",
    markers=True,
    title="Évolution du bilan de masse cumulée des glaciers",
    labels={
        "Year": "Année",
        "Mean cumulative mass balance": "Bilan de masse cumulée (m équiv. eau)"
    }
)

fig.update_layout(
    margin={"r":0, "t":40, "l":0, "b":0},
    plot_bgcolor="white"
)

fig.show()
```

```python
# Bilan de masse cumulée et observations
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df["Year"],
    y=df["Mean cumulative mass balance"],
    mode="lines+markers",
    name="Bilan de masse cumulée"
))

fig.add_trace(go.Bar(
    x=df["Year"],
    y=df["Number of observations"],
    name="Nombre d'observations",
    opacity=0.4,
    yaxis="y2"
))

fig.update_layout(
    title="Glaciers : bilan de masse cumulée et observations",
    yaxis=dict(title="Bilan de masse (m)"),
    yaxis2=dict(
        title="Nombre d'observations",
        overlaying="y",
        side="right"
    ),
    margin={"r":0, "t":40, "l":0, "b":0}
)

fig.show()
```

```python
import plotly.express as px

fig = px.line(
    df,
    x="Year",
    y="Mean cumulative mass balance",
    markers=True,
    title="Évolution du bilan massique moyen cumulatif des glaciers"
)

fig.update_layout(
    xaxis_title="Année",
    yaxis_title="Bilan massique cumulatif (mwe : mètres équivalent en eau)",  
    template="plotly_white"
)

fig.show()
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x="Year",
    y="Mean cumulative mass balance",
    title="Dispersion annuelle du bilan massique cumulatif des glaciers"
)

fig.update_layout(
    xaxis_title="Année",
    yaxis_title="Bilan massique cumulatif",
    template="plotly_white"
)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df["Year"],
    y=df["Mean cumulative mass balance"],
    name="Bilan massique",
    mode="lines+markers"
))

fig.add_trace(go.Scatter(
    x=df["Year"],
    y=df["Number of observations"],
    name="Nb d'observations",
    mode="lines",
    yaxis="y2"
))

fig.update_layout(
    title="Évolution du bilan massique et du nombre d’observations",
    xaxis=dict(title="Année"),
    yaxis=dict(title="Bilan massique"),
    yaxis2=dict(title="Nb observations", overlaying="y", side="right"),
    template="plotly_white"
)

fig.show()
```

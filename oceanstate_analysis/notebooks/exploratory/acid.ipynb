```python
df.dropna()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 6)
plt.figure(figsize=(12, 8))

# Graphique linéaire de l'évolution de l'acidification
plt.subplot(2, 1, 1)
plt.plot(df['year'], df['Ocean_acidification(in_PH)'],
         marker='o', linewidth=2, markersize=4, color='steelblue')
plt.title('Évolution de l\'acidification océanique (pH) de 1950 à 2023',
          fontsize=14, fontweight='bold')
plt.xlabel('Année')
plt.ylabel('pH océanique')
plt.grid(True, alpha=0.3)


plt.tight_layout()
plt.show()
```
